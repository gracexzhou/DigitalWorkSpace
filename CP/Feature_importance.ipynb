```python
#prepare the feature list name column - step 1 of 2

feature_columns_ascending = []
feature_imp_values_ascending= []

for value in feature_list_ascending:
  feature_columns_ascending.append(value[0])
  feature_imp_values_ascending.append(value[1])
```

```python
#prepare the feature list name column - step 2 of 2

import re
feature_columns_cut = []

for str in feature_columns_ascending:
    pattern = r"_kurtosis|_sum|_var|_avg|_max"
    feature_columns_cut.append(re.sub(pattern,"",str))
print(feature_columns_cut)

```

```python
#combine the updated feature name column with its feature value
feature_dic_updated = dict(zip(feature_columns_cut,feature_imp_values_ascending))
print(feature_dic_updated)

#convert dict to list
feature_list_ascending_updated = list(feature_dic_updated.items())
print(feature_list_ascending_updated)
```

```python
#sort the updated list to desending
feature_list_desending = sorted(feature_list_ascending_updated, key=lambda x: x[1], reverse=True)
print(feature_list_desending)
```

```python
#for py chart
feature_columns_desending = []
feature_imp_values_desending = []

for value in feature_list_desending:
  feature_columns_desending.append(value[0])
  feature_imp_values_desending.append(value[1])
```

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 8), dpi=80)
plt.barh(feature_columns_desending[:10], feature_imp_values_desending[:10])
plt.ylabel("Important Feature Columns")
plt.xlabel("Importance Score")
```
